```python
plt.figure(figsize=(10,6))
plt.title("Overall cereals data distribution")
sns.histplot(data=df)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['mfr'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Manufacturer')
plt.xlabel('mfr')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['type'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Cold or Hot Types of cereal')
plt.xlabel('type')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['calories'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Calories')
plt.xlabel('calories')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['protein'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Protein')
plt.xlabel('protein')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['fat'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of fat')
plt.xlabel('fat')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['sodium'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Sodium')
plt.xlabel('sodium')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Cereal Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='calories', y='rating', data=df, color='orange')
plt.title('Relationship between Calories and Ratings')
plt.xlabel('Calories')
plt.ylabel('Rating')
plt.show
```

```python
cereal_sorted = df.sort_values('rating', ascending=False)
cereal_sorted
```

```python
plt.figure(figsize=(15,7))
plt.title("Brands and Their Ratings")
plt.xticks(rotation=90)
sns.barplot(data=cereal_sorted, x=cereal_sorted['name'], y=cereal_sorted['rating'])
```

```python
plt.figure(figsize=(10,6))
plt.title("Relationship between sugar content and Rating")
plt.xticks(rotation=90)
sns.regplot(data=df, x=df['sugars'], y=df['rating'])
```
